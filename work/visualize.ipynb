```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gnss = pd.read_csv("/home/nakao/AdaptiveFEM/work/work_1250_1/data/GNSS_reduced.dat", delim_whitespace=True)
displacement = np.loadtxt("/home/nakao/AdaptiveFEM/work/work_1250_1/data/greens_function.dat")
ux = []
uy = []
uz = []
for i in range(displacement.shape[0]//3):
    ux.append(displacement[i*3])
    uy.append(displacement[i*3+1])
    uz.append(displacement[i*3+2])
ux = np.array(ux)
uy = np.array(uy)
uz = np.array(uz)
```
